```python
import os
from ultralytics import YOLO
```

```python

os.environ['WANDB_MODE'] = 'disabled'
os.environ['WANDB_DISABLED'] = 'true'
```

```python

model = YOLO('yolov8n.pt')
results = model.train(
    data='/kaggle/input/coco-dataset-for-finetune-yolo-v8/dataset/data.yaml',
    epochs=100,
    imgsz=640
    batch=16,
    name='my_yolo_model',
    project='model'
)
```
